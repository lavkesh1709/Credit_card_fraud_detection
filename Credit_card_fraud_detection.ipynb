```python
ratio = 0.75

total_rows = bal_dataset.shape[0]
train_size = int(total_rows*ratio)

train_df = bal_dataset[0:train_size]
test_df = bal_dataset[train_size:]
```

```python

x_train = np.array(train_df.iloc[:,0:30])
x_test = np.array(test_df.iloc[:,0:30])
y_train = np.array(train_df.iloc[:,30])
y_test = np.array(test_df.iloc[:,30])
```

```python
def zscore_norm(x):
    mean = np.mean(x,axis = 0)
    dev = np.std(x,axis = 0)
    x_norm = (x-mean)/dev
    return x_norm
```

```python
xtrain_norm = zscore_norm(x_train)
xtest_norm = zscore_norm(x_test)
```

```python
m,n = xtrain_norm.shape
```

```python
def sigmoid(z):

    g = 1/(1+np.exp(-z))
    return g
```

```python

w = np.zeros(n)
b = 0.0
```

```python
def cost(x,y,w,b):
    loss = 0.0
    for i in range(m):
        z_i = (np.dot(x[i],w)+b)
        f_wb_i = sigmoid(z_i)
        loss += (-y[i]*np.log(f_wb_i)-(1-y[i])*(np.log(1-f_wb_i)))
    cost = loss/m
    return cost
```

```python
cost(xtrain_norm,y_train,w,b)
```

```python
def gradient(x,y,w,b):
    dj_dw = np.zeros(n)
    dj_db = 0
    for i in range(m):
        z_i = (np.dot(x[i],w)+b)
        f_wb_i = sigmoid(z_i)
        err_i = (f_wb_i-y[i])
        dj_db += err_i
        for j in range(n):
            dj_dw[j] += err_i*x[i,j]
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w,b,cost,gradient,alpha,epochs):
    for i in range(epochs):
        dj_dw,dj_db = gradient(x,y,w,b)
        w = w - alpha*dj_dw
        b = b - alpha*dj_db

        if i%100 == 0:
            print(f"Cost after {i}th iteration: {cost(x,y,w,b)}")
    return w,b
```

```python
w,b = gradient_descent(xtrain_norm,y_train,w,b,cost,gradient,0.2,1000)
```

```python
z = np.dot(xtest_norm,w)+b
y_test_pred =  1/(1+np.exp(-z))
```

```python
ytest_pred = np.where(y_test_pred > 0.5, 1, 0)
```

```python
def f1_score(y_true, y_pred):
    true_positives = np.sum((y_true == 1) & (y_pred == 1))
    true_negatives = np.sum((y_true == 0) & (y_pred == 0))
    false_positives = np.sum((y_true == 0) & (y_pred == 1))
    false_negatives = np.sum((y_true == 1) & (y_pred == 0))
    accuracy=(true_positives +true_negatives)/(true_positives+true_negatives+false_positives +false_negatives)

    precision = true_positives / (true_positives + false_positives)
    recall = true_positives / (true_positives + false_negatives)

    score = 2 * (precision * recall) / (precision + recall)
    return score,accuracy,precision,recall
```

```python
f1_Score,Accuracy,Precision,Recall = f1_score(y_test,ytest_pred)
```
